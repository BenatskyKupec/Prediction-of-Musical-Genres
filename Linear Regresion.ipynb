```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
```

```python
path = "less_genres.csv"

data = pd.read_csv(path)
```

```python
# get rid of unusefull columns: track_id, time_signature
# get rid of highly corelated columns: loudness, acousticness, valence

data = data.drop(['track_id', 'time_signature', 'loudness', 'acousticness', 'valence'], axis=1)
data = data.drop(['artist_name', 'track_name'], axis=1)
```

```python
data
```

```python
#create train and test
Y = data['genre']
X = data.drop('genre', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.4, random_state=42)
```

```python
# stupidly (1...num of genres) encode genres

y_train = y_train.replace(to_replace=pd.unique(y_train), value=np.arange(0, y_train.nunique()))
y_test = y_test.replace(to_replace=pd.unique(y_test), value=np.arange(0, y_test.nunique()))

pass        # to omit FutureWarning
```

```python
# DOES NOT IMPROVE

scaler = preprocessing.StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)

print(lm.intercept_, lm.coef_)
```

```python
z = (lm.predict(X_train).round() == y_train).value_counts()
print("Accuracy on train: ", z[True]/(z[False] + z[False]))

z = (lm.predict(X_test).round() == y_test).value_counts()
print("Accuracy on test: ", z[True]/(z[False] + z[False]))
```
