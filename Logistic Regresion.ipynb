```python
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
# uncomment rows below to run on original genres

path = "less_genres.csv"
# path = 'spotify_data.csv'


data = pd.read_csv(path)
# data = data.drop('Unnamed: 0', axis=1)
```

```python
data = data.drop(['track_id', 'loudness'], axis=1)
# do not think it would have much effect and the PC has very little RAM to one-hot-encode :{
data = data.drop(['artist_name', 'track_name'], axis=1)
```

```python
data
```

```python
# divide data into train and test
Y = data['genre']
X = data.drop('genre', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.4, random_state=42)
```

```python
# scaling

scaler = preprocessing.StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
model_logReg = LogisticRegression(random_state=42)
model_logReg.fit(X_train, y_train)

preds = model_logReg.predict(X_test)
```

```python
print(classification_report(y_test, preds))
```
